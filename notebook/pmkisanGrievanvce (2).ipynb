#ENCRYPT

```python
from Crypto.Cipher import AES
import base64

# --- Static Keys ---
key_hex = "B275D03C722F85941A287A08B27EA7C70BE436D9BC85A7EB60BFB53AFEA273C6"
iv_hex  = "62958328A844AAE69BDFFBFC7F3D9E1C"

key = bytes.fromhex(key_hex)
iv  = bytes.fromhex(iv_hex)

# --- JSON payload to encrypt ---
import json
json_payload ={
    "Type": "IdentityNo_Details",
    "TokenNo": "PMK_123456",
    "IdentityNo": "b23907d5-4bc8-3f2b-9549-bac8a4c9f760",
    "GrievanceType": "G002",
    "GrievanceDescription": "Test Grievance for VISTAAR based on Aadhar"
}
plaintext = json.dumps(json_payload).encode()

# --- Encrypt ---
cipher = AES.new(key, AES.MODE_GCM, nonce=iv)
ciphertext, tag = cipher.encrypt_and_digest(plaintext)

# AES-GCM output: ciphertext || tag
encrypted_b64 = base64.b64encode(ciphertext + tag).decode()
print("EncryptedRequest JSON:")
print(json.dumps({"EncryptedRequest": encrypted_b64}, indent=2))
```
